```python
import pandas as pd
from datetime import datetime
import calendar
```

```python
data = pd.read_csv('bike_rental_raw.csv')
data
```

```python
data.columns
```

```python
data['season'] = data['season'].map({1: 'Spring',
                                      2: 'Summer',
                                      3: 'Fall',
                                      4: 'Winter'})
data.head()
```

```python
data['weather'] = data['weather'].map({1: 'Clear',
                                         2: 'Mist, Few clouds',
                                         3: 'Light snow, Rain, Thunderstorm',
                                         4: 'Heavy Rain, Thunderstorm, Snow, Fog'})
data.head()
```

```python
data.weather.unique()
```

```python
data.to_csv('bike_rental.csv', index=False)
```
